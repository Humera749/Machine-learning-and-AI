```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.preprocessing import StandardScaler
from scipy.stats import norm
from scipy.optimize import minimize

# Given data
X = X
Y = Y

# Scale both X and Y
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_Y = StandardScaler()
Y_scaled = scaler_Y.fit_transform(Y.reshape(-1, 1)).ravel()

# Define the kernel and Gaussian Process model with increased max_iter
kernel = C(1.0, (1e-3, 1e3)) * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=5, normalize_y=True)

# Fit to the scaled data
gp.fit(X_scaled, Y_scaled)

# Define the acquisition function (Expected Improvement)
def acquisition_function(x, gp, y_max):
    mean, std = gp.predict(x.reshape(1, -1), return_std=True)
    mean, std = mean[0], std[0]
    
    Z = (mean - y_max) / std
    ei = (mean - y_max) * norm.cdf(Z) + std * norm.pdf(Z)
    return -ei  # Negative for minimization

# Optimization loop
n_iterations = 25
y_max = max(Y_scaled)
new_points = []

for i in range(n_iterations):
    res = minimize(acquisition_function, np.random.rand(4), args=(gp, y_max),
                   bounds=[(0, 1)]*4, method='Powell')
    
    # Predict yield at the new point
    x_next_scaled = scaler_X.transform(res.x.reshape(1, -1))
    y_next = gp.predict(x_next_scaled)[0]
    
    # Add new point
    X_scaled = np.vstack((X_scaled, x_next_scaled))
    Y_scaled = np.append(Y_scaled, y_next)
    new_points.append((res.x, scaler_Y.inverse_transform(y_next.reshape(-1, 1)).item()))
    
    # Update model
    gp.fit(X_scaled, Y_scaled)
    y_max = max(y_max, y_next)

# Final result display
print("Estimated optimal yield (unscaled):", scaler_Y.inverse_transform([[y_max]])[0, 0])
optimal_index = np.argmax(Y_scaled)
print("Optimal chemical combination (scaled):", scaler_X.inverse_transform(X_scaled[optimal_index].reshape(1, -1)).ravel())

# Plot convergence of yield
y_values = [point[1] for point in new_points]
plt.plot(range(1, len(y_values) + 1), y_values, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Predicted Yield")
plt.title("Convergence of Gaussian Process Optimization")
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Given data
X = X
Y = Y

# Scale inputs and outputs
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_Y = StandardScaler()
Y_scaled = scaler_Y.fit_transform(Y.reshape(-1, 1)).ravel()

# Define Polynomial Regression Model
poly = PolynomialFeatures(degree=3)  # Use degree 2 or 3 for complexity control
X_poly = poly.fit_transform(X_scaled)

# Fit Polynomial Regression
model = LinearRegression()
model.fit(X_poly, Y_scaled)

# Define the function to maximize (negative because we minimize for maximizing)
def objective(x):
    x = x.reshape(1, -1)
    x_poly = poly.transform(scaler_X.transform(x))
    y_pred = model.predict(x_poly)
    return -y_pred[0]  # Negative for maximization

# Optimization settings
initial_guess = np.random.rand(4)
bounds = [(0, 1)] * 4  # Since our inputs are within [0, 1]

# Run optimization
result = minimize(objective, initial_guess, bounds=bounds, method="L-BFGS-B")
optimal_input_scaled = result.x
optimal_input = scaler_X.inverse_transform(optimal_input_scaled.reshape(1, -1))

# Predict the maximum yield
optimal_yield_scaled = model.predict(poly.transform([optimal_input_scaled]))[0]
optimal_yield = scaler_Y.inverse_transform([[optimal_yield_scaled]])[0, 0]

print("Optimal chemical combination (unscaled):", optimal_input)
print("Predicted maximum yield:", optimal_yield)

# Plot convergence of predicted yields
n_iterations = 200
y_values = []

for _ in range(n_iterations):
    res = minimize(objective, np.random.rand(4), bounds=bounds, method="L-BFGS-B")
    y_pred = -objective(res.x)  # Convert back to positive for yield
    y_values.append(scaler_Y.inverse_transform([[y_pred]]).item())

plt.plot(range(1, len(y_values) + 1), y_values, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Predicted Yield")
plt.title("Convergence of Polynomial Regression Optimization")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ranges of each variable from 0 to 1
x_range = np.linspace(0, 1, 100)

# Set the fixed values for all other inputs (e.g., mean values of each column)
fixed_values = X.mean(axis=0)

# Prepare subplots for each input variable
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for i in range(4):  # Loop over each input variable
    y_pred = []  # Store predictions for this input variable

    for x_val in x_range:
        # Create an input array with the fixed values, changing only the i-th variable
        x_input = fixed_values.copy()
        x_input[i] = x_val  # Update only the current variable

        # Scale input, transform with polynomial features, and predict yield
        x_input_scaled = scaler_X.transform([x_input])  # Scale the input
        x_poly = poly.transform(x_input_scaled)  # Apply polynomial transformation
        y_scaled_pred = model.predict(x_poly)  # Predict scaled output
        y_pred.append(scaler_Y.inverse_transform([[y_scaled_pred[0]]])[0, 0])  # Unscale prediction

    # Plot yield vs. the varying input variable
    axes[i].plot(x_range, y_pred, label=f"Variable {i + 1}")
    axes[i].set_xlabel(f"Variable {i + 1} Value")
    axes[i].set_ylabel("Predicted Yield")
    axes[i].set_title(f"Effect of Variable {i + 1} on Yield")
    axes[i].legend()

plt.tight_layout()
plt.show()
```
