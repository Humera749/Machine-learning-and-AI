```python
import matplotlib.pyplot as plt
from skopt import gp_minimize
from skopt.space import Real
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Input (X) and Output (Y) arrays
X = X
Y = Y

# Define the model (RandomForestRegressor)
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Define the objective function
def objective(params):
    # Set the hyperparameters of the model based on the parameters from optimization
    model.set_params(max_depth=int(params[0] * 10), min_samples_split=int(params[1] * 5),
                     max_features=params[2], min_samples_leaf=int(params[3] * 2),
                     n_estimators=int(params[4] * 100), random_state=42)

    # Fit the model on the data
    model.fit(X, Y)

    # Use Mean Squared Error (MSE) as the metric to minimize
    pred = model.predict(X)
    mse = mean_squared_error(Y, pred)
    
    return mse

# Define the search space for the hyperparameters
space = [
    Real(1, 10, name='max_depth'),
    Real(2, 10, name='min_samples_split'),
    Real(0.1, 1.0, name='max_features'),
    Real(1, 10, name='min_samples_leaf'),
    Real(0.1, 1.0, name='n_estimators'),
    Real(0.1, 1.0, name='random_state')
]

# Perform Bayesian Optimization
result = gp_minimize(
    objective,
    space,
    n_calls=50,  # Number of function evaluations
    n_initial_points=10,  # Initial random evaluations
    acq_func="EI",  # Acquisition function: Expected Improvement
    random_state=42
)

# Print the best found hyperparameters
print("Best hyperparameters:", result.x)
print("Best score (MSE):", result.fun)

# Plot the convergence
plt.figure(figsize=(8, 6))
plt.plot(range(len(result.func_vals)), result.func_vals, marker='o', color='b', linestyle='-', markersize=5)
plt.xlabel('Number of Function Evaluations')
plt.ylabel('Objective Function Value (MSE)')
plt.title('Convergence Plot of Bayesian Optimization')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skopt import gp_minimize
from skopt.plots import plot_convergence
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.gaussian_process.kernels import Matern, ConstantKernel as C

# Your input (X) and output (Y) arrays
X = X
Y = Y

# Ensure X and Y are correctly shaped
print(X.shape, Y.shape)  # X should have shape (n_samples, 6)

# Define the search space for 6 features
space = [Real(0, 1) for _ in range(6)]

# Define the kernel
kernel = C(1.0, (1e-2, 1e2)) * Matern(length_scale=1.0, nu=2.5)

# Create and fit the Gaussian Process model
gp = GaussianProcessRegressor(kernel=kernel, random_state=42)
gp.fit(X, Y)

# Define the objective function
def objective(X_new):
    X_new = np.array(X_new).reshape(1, -1)  # Reshape to (1, 6)
    predicted = gp.predict(X_new)
    return -predicted[0]

# Perform the optimization
result = gp_minimize(objective, space, n_calls=50, acq_func="LCB", random_state=42, n_jobs=1)

# Plot the convergence plot
plot_convergence(result)
plt.title("Convergence Plot for Maximization")
plt.show()

# Print the best parameters and value found
print(f"Best found parameters: {result.x}")
print(f"Best value found: {-result.fun}")
```
